```python
import matplotlib.pyplot as plt #accuracy
import numpy as np #imageto array
import pandas as pd
import seaborn as sns
import os #fileprocess

# Importing Deep Learning Libraries

from tensorflow.keras.preprocessing.image import load_img, img_to_array
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Dense,Input,Dropout,GlobalAveragePooling2D,Flatten,Conv2D,BatchNormalization,Activation,MaxPooling2D
from keras.models import Model,Sequential
from tensorflow.keras.optimizers import Adam,SGD,RMSprop


picture_size = 224


batch_size  = 32

datagen_train  = ImageDataGenerator() #class
datagen_val = ImageDataGenerator()

train_set = datagen_train.flow_from_directory("C:/Users/Dell/Desktop/Untitled Folder 1/train",
                                              target_size = (picture_size,picture_size),
                                              color_mode = "rgb",
                                              batch_size=batch_size,
                                              class_mode='categorical')


test_set = datagen_val.flow_from_directory("C:/Users/Dell/Desktop/Untitled Folder 1/test",
                                              target_size = (picture_size,picture_size),
                                              color_mode = "rgb",
                                              batch_size=batch_size,
                                              class_mode='categorical')
```

```python
import keras
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
no_of_classes = 2

#model archi

model = ResNet50(
    input_shape = (224, 224, 3), # Making the image into 3 Channel, so concating 3.
    weights = 'imagenet', # Default weights.
    include_top = False   # 
)
    # mark loaded layers as not trainable

# add new classifier layers
flat1 = Flatten()(model.layers[-1].output)

output = Dense(2, activation='softmax')(flat1)
# define new model
model = Model(inputs=model.inputs, outputs=output)
# compile model

model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

#save model after every epoch and load to work for other epoch
checkpoint = ModelCheckpoint("resnet.h5", monitor='val_acc', verbose=1, save_best_only=True, mode='max')

#callbacks_list = [early_stopping,checkpoint,reduce_learningrate]


#common classi(adam)

model.compile(loss='categorical_crossentropy',
              optimizer = Adam(learning_rate=0.001),
              metrics=['accuracy'])
```

```python
history_object = model.fit_generator(train_set, 
                                     steps_per_epoch=18,
                                      validation_data=test_set,
                                      validation_steps=3, 
                                     epochs=100, verbose=1)
model.save("C:/Users/Dell/Desktop/Untitled Folder 1/sample1.h5")
```
